```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Vamos supor que 'df_Tupã' seja o DataFrame que contém os dados do ENEM

# Plotagem do gráfico
plt.figure(figsize=(10, 6))
plt.scatter(df_Tupã['IDADE'], df_Tupã['NOTA_REDACAO'], alpha=0.5)
plt.title('Nota da Redação em função da Idade')
plt.xlabel('Idade')
plt.ylabel('Nota da Redação')
plt.grid(True)
plt.show()
```

```python
# Supondo que 'df_Tupã' seja o DataFrame que contém os dados do ENEM

# Calculando a média da nota da redação para cada idade
mean_notas_por_idade = df_Tupã.groupby('IDADE')['NOTA_REDACAO'].mean()

# Plotando o gráfico de linhas
plt.figure(figsize=(10, 6))
mean_notas_por_idade.plot(kind='line')
plt.title('Média da Nota da Redação por Idade')
plt.xlabel('Idade')
plt.ylabel('Média da Nota da Redação')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Supondo que 'df_Tupã' seja o DataFrame que contém os dados do ENEM

# Calculando a média da nota da redação para cada idade
mean_notas_por_idade = df_Tupã.groupby('IDADE')['NOTA_REDACAO'].mean().reset_index()

# Criando o gráfico de linha interativo
fig = px.line(mean_notas_por_idade, x='IDADE', y='NOTA_REDACAO', title='Média da Nota da Redação por Idade',
labels={'IDADE': 'Idade', 'NOTA_REDACAO': 'Média da Nota da Redação'})
```
