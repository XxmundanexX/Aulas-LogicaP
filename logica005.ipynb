```python
import pandas as pd
import numpy as np
```

```python
tabela = pd.read_csv("Dataset_AulaPratica_NOVO.csv", sep = ";", encoding = "iso 8859-1")
```

```python
tabela
```

```python
# Remover as colunas 'Unnamed: 7' até 'Unnamed: 12'
tabela = tabela.drop(tabela.columns[7:13], axis=1)

# Remover as linhas com id >= 44
tabela = tabela[tabela['ID'] < 44]
```

```python
tabela
```

```python
# Calcular o valor médio da variável Altura(cm)
valor_medio_altura = tabela['Altura(cm)'].mean()

print(f"O valor médio da variável Altura(cm) é: {valor_medio_altura:.2f}")
```

```python
# Calcular o valor médio da variável idade
media_idade = tabela['Idade'].mean()

print(f"O valor médio da variável idade é: {media_idade:.0f}")
```

```python
 # Calcular o valor médio da variável idade
media_calcado = tabela["Tam. Calçado"].mean()

print(f"O valor médio da variável idade é: {media_calcado:.0f}")
```

```python
# Calcular o IMC (Índice de Massa Corporal)
tabela['IMC'] = tabela['Peso(kg)'] / ((tabela['Altura(cm)'] / 100) ** 2)
```

```python
tabela["IMC"]
```

```python
def classificar_imc(imc):
    if imc < 18.5:
        return 'Abaixo do peso'
    elif 18.5 <= imc < 25:
        return 'Peso normal'
    elif 25 <= imc < 30:
        return 'Sobrepeso'
    elif 30 <= imc < 35:
        return 'Obesidade Grau I'
    elif 35 <= imc < 40:
        return 'Obesidade Grau II (severa)'
    else:
        return 'Obesidade Grau III (mórbida)'

# Adicionando uma nova coluna com a classificação de IMC
tabela['Classificação_IMC'] = tabela['IMC'].apply(classificar_imc)

# Exibindo as primeiras linhas do DataFrame com a nova coluna
print(tabela[['IMC', 'Classificação_IMC']])
```
